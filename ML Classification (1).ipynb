```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
df = pd.DataFrame(X, columns=data.feature_names)
```

```python
#checking  missing value 
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# train-test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# initialize
log_reg = LogisticRegression(max_iter=10000)
```

```python
# train_model
log_reg.fit(X_train, y_train)
```

```python
y_pred_log_reg = log_reg.predict(X_test)
log_reg_accuracy = accuracy_score(y_test, y_pred_log_reg)
log_reg_accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# initialize
dt_classifier = DecisionTreeClassifier(random_state=42)
```

```python
dt_classifier.fit(X_train, y_train)
```

```python
y_pred_dt = dt_classifier.predict(X_test)
dt_accuracy = accuracy_score(y_test, y_pred_dt)
dt_accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#initialize
rf_classifier = RandomForestClassifier(random_state=42)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python
y_pred_rt = rf_classifier.predict(X_test) 
rf_accuracy = accuracy_score(y_test, y_pred_rt)
rf_accuracy
```

```python
from sklearn.svm import SVC
```

```python
#initialize
svm_classifier = SVC(random_state=42)
```

```python
#train the Model 
svm_classifier.fit(X_train, y_train)
```

```python
y_pred_svm = svm_classifier.predict(X_test)
svm_accuracy = accuracy_score(y_test, y_pred_svm)
svm_accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#initialize
k_nn_classifier = KNeighborsClassifier()
```

```python
#train the model
k_nn_classifier.fit(X_train, y_train)
```

```python
y_pred_k_nn= k_nn_classifier.predict(X_test)
k_nn_accuracy = accuracy_score(y_test, y_pred_k_nn)
k_nn_accuracy
```

```python
# comparison all five classification algorithms
results = {
    'Logistic Regression':log_reg_accuracy,
    'Decision Tree':dt_accuracy,
    'Random Forest':rf_accuracy,
    'SVM': svm_accuracy,
    'k-NN':k_nn_accuracy
}
```

```python
results = sorted(results.items(), key=lambda x: x[1], reverse=True)
results
```
